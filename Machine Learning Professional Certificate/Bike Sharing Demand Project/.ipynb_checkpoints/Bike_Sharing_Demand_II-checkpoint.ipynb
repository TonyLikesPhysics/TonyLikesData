### The Objective and Planning

The main objective of this analysis is to get a working comparison of a linear regression model, a Ridge regression model, and a Lasso regression model. To accomplish this task, pipelines are employed along with GridSearchCV so scaling, modeling, hyperparameter tuning and cross validation can all be done in one step.

The plan is to do some initial EDA as in the previous project, but we will cut the categorical columns and just look at the numeric columns for now. Additionally, one of the columns representing temperature is dropped to avoid multicollinearity and the models are run using only the humidity, windspeed, temp features and count as the target.

```python
bike_data = data[['temp', 'atemp', 'humidity','windspeed','count']]
```

```python
bike_data.head()
```

```python
sns.heatmap(bike_data.corr(),annot=True)
```

```python
bike_data.drop('atemp', axis=1, inplace=True)
```

### Linear Regression Model

```python
y_col = "count"

X = bike_data.drop(y_col, axis=1)
y = bike_data[y_col]
```

```python
kf = KFold(shuffle=True, random_state=72018, n_splits=3)
```

```python
estimator = Pipeline([("scaler", StandardScaler()),
        ("polynomial_features", PolynomialFeatures()),
        ("regression", LinearRegression())])

#estimator.get_params().keys()

params = {
    'polynomial_features__degree': [1, 2, 3]
}

grid = GridSearchCV(estimator, params, cv=kf)

```

```python
grid.fit(X,y)
```

```python
grid.best_score_, grid.best_params_
```

```python
y_predict = grid.predict(X)
```

```python
r2_score(y, y_predict)
```

### Using Ridge Regression

```python

estimator = Pipeline([("scaler", StandardScaler()),
        ("polynomial_features", PolynomialFeatures()),
        ("ridge_regression", Ridge())])

params = {
    'polynomial_features__degree': [1, 2, 3],
    'ridge_regression__alpha': np.geomspace(1e-7, 1e2, num=10)
}

grid1 = GridSearchCV(estimator, params, cv=kf)
```

```python
grid1.fit(X,y)
```

```python
grid1.best_score_, grid1.best_params_
```

```python
y_predict1 = grid1.predict(X)
```

```python
# This includes both in-sample and out-of-sample
r2_score(y, y_predict1)
```

### Using Lasso Regression

```python
estimator = Pipeline([("scaler", StandardScaler()),
        ("polynomial_features", PolynomialFeatures()),
        ("lasso_regression", Lasso())])

params = {
    'polynomial_features__degree': [1, 2, 3],
    'lasso_regression__alpha': np.geomspace(1e-7, 1e2, num=10)
}

grid2 = GridSearchCV(estimator, params, cv=kf)
```

```python
grid2.fit(X,y)
```

```python
grid2.best_score_, grid2.best_params_
```

```python
y_predict2 = grid2.predict(X)
```

```python
# This includes both in-sample and out-of-sample
r2_score(y, y_predict2)
```

### Key Findings and Flaws

In my inital model shown above, I was not able to get the R^2 value above 0.28. Suprisingly, all three models are close to withing a small fraction, however the regular linear regression model actually does perform a very small fraction (0.00001 to 0.00002) better than the other models. I suspect that the categorical features that were removed were of much greater importance. The next step in this process is to see if I can add these features back in, rename the ambiguous values in the columns that were removed, and then use one-hot encoding before running the regressions. Getting the R^2 value up above at least around 60% is my goal for further work on this project.
