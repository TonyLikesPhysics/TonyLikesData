### Problem 1
---

Multiplying out $\omega_4$ by the input matrix $\alpha$ gives us the corresponding output matrix $\beta$. 

$\big|\beta\big\rangle =
\begin{bmatrix}
1 & 1 & 1 & 1\\
1 & i & -1 & -i\\
1 & -1 & 1 & -1\\
1 & -i & -1 & i\\
\end{bmatrix}\begin{bmatrix}
\alpha_0\\
\alpha_1\\
\alpha_2\\
\alpha_3\\
\end{bmatrix}
$ = $
\begin{bmatrix}
\alpha_0+\alpha_1+\alpha_2+\alpha_3\\
\alpha_0+i\alpha_1-\alpha_2-i\alpha_3\\
\alpha_0-\alpha_1+\alpha_2-\alpha_3\\
\alpha_0-i\alpha_1-\alpha_2+i\alpha_3\\
\end{bmatrix}\begin{bmatrix}
\beta_0\\
\beta_1\\
\beta_2\\
\beta_3\\
\end{bmatrix}
$

$ \big|\beta'\big\rangle =
\begin{bmatrix}
1 & 1 & 1 & 1\\
1 & i & -1 & -i\\
1 & -1 & 1 & -1\\
1 & -i & -1 & i\\
\end{bmatrix}\begin{bmatrix}
\alpha_1\\
\alpha_2\\
\alpha_3\\
\alpha_0\\
\end{bmatrix}
$ = $
\begin{bmatrix}
\alpha_0+\alpha_1+\alpha_2+\alpha_3\\
-i\alpha_0+\alpha_1+i\alpha_2-\alpha_3\\
-\alpha_0+\alpha_1-\alpha_2+\alpha_3\\
i\alpha_0+\alpha_1-i\alpha_2-\alpha_3\\
\end{bmatrix}\begin{bmatrix}
\beta_0'\\
-i\beta_1'\\
-\beta_2'\\
i\beta_3'\\
\end{bmatrix}
$

By cycling the input we can see that the output matrix $\beta'$ is only different from $\beta$ by a relative phase shift. So it is proven that for each element $\beta_j$, the probability $|\beta_j^2|$ is equal to that of $|\beta_j'^2|$.

### Problem 3

```python
# Problem 3.1 - Construct the 2-quibit Uf oracle for the function f(x) with period 1
oracle1 = QuantumCircuit(2, name = 'Uf1 Period 1')
oracle1.h([0,1])
oracle1.draw('mpl')
```

```python
oracle1.to_gate()
```

```python
# Problem 3.2 Apply QFT to the period 1 oracle
qcircuit1 = QuantumCircuit(2)
qcircuit1.append(oracle1, [0,1])
qcircuit1.h(1)
qcircuit1.cp(pi/2, 0,1)
qcircuit1.h(0)
qcircuit1.swap(0,1)
qcircuit1.draw('mpl')
```

```python
# Problem 3.3 Probing the statevector
job1 = execute(qcircuit1, backend, shots =1)
result1 = job1.result()
sv1 = result1.get_statevector()
np.around(sv1,2)
```

```python
plot_bloch_multivector(sv1)
```

I decided to probe the circuit using the statevector simulator to see if I could find the period using this method. I this situation it seems to work. Printing the state vector array shows us that only the $\big|00\big\rangle$ basis state is output, and the Bloch vector plot shows that both qubits are in the up state. 

---

```python
# Problem 3.4 Taking measurements on the 2-qubit Period 1 
qcircuit1 = QuantumCircuit(2,2)
qcircuit1.append(oracle1, [0,1])
qcircuit1.h(1)
qcircuit1.cp(pi/2, 0,1)
qcircuit1.h(0)
qcircuit1.swap(0,1)
qcircuit1.measure([0,1],[0,1])
#qcircuit1.measure(0,0)
qcircuit1.draw('mpl')
```

```python
# Problem 3.5 Measuring the probability using the qasm simulator
resultq1 = execute(qcircuit1, backend = simulator, shots = 1024).result()
counts1 = resultq1.get_counts()
print(counts1)
plot_histogram(counts1, legend = ['counts'])
```

As expected for the $U_f$ of period 1, we see that 100% of the counts fall in the $\beta_0$ basis state, which is $\big|00\big\rangle$.
