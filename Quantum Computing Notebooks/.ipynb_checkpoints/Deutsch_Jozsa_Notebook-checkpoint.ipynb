### Problem 4
---

### 4.1 - Statevector Simulation - Deutsch Circuit - Oracle 1 

```python
# Problem 4.1
job1 = execute(deutsch_circuit1, backend, shots =1)
result1 = job1.result()
sv1 = result1.get_statevector()
np.around(sv1,2)
#print(sv1)
```

```python
plot_bloch_multivector(sv1)
```

When running the statevector simulation without taking measurements, we can see that the first qubit is in a definite state depending on whether the oracle we used is balanced or constant. The first oracle is constant and we get the following statevector:

$$
\frac{1}{\sqrt{2}}\begin{bmatrix}1&0\\-1&0\end{bmatrix}=\frac{1}{\sqrt{2}}\Big(\big|00\big\rangle-\big|10\big\rangle\Big)=\frac{1}{\sqrt{2}}\Big(\big|0\big\rangle-\big|1\big\rangle\Big)\big|0\big\rangle
$$

We can see in the bloch vector visualization that qubit 0's statevector is in the input state $\big|0\big\rangle$ but qubit 1's statevector has been rotated to point in the $\big|0\big\rangle-\big|1\big\rangle$ direction, which is true for all four circuit cases, so we can see why when we measure this qubit that 50% of the time it will be measured in the up direction and the other 50% it'll be measured in down. This is what we should expect of the Deutsch alogrithm and it illustrates how we can probe whether the circuit is balanced or constant without taking a measurement and we only need one shot to do it.

### 4.2 - Statevector Simulation - Deutsch Circuit - Oracle 2

```python
# Problem 4.2 
job2 = execute(deutsch_circuit2, backend, shots =1)
result2 = job2.result()
sv2 = result2.get_statevector()
np.around(sv2,2)
```

```python
plot_bloch_multivector(sv2)
```

The second oracle is balanced and we get the following statevector:

$$
\frac{1}{\sqrt{2}}\begin{bmatrix}0&1\\0&-1\end{bmatrix}=\frac{1}{\sqrt{2}}\Big(\big|01\big\rangle-\big|11\big\rangle\Big)=\frac{1}{\sqrt{2}}\Big(\big|0\big\rangle-\big|1\big\rangle\Big)\big|1\big\rangle
$$

We can see in the bloch vector visualization that qubit 0's state has been flipped from the input state $\big|0\big\rangle$ to the state $\big|1\big\rangle$.

### 4.3 - Statevector Simulation - Deutsch Circuit - Oracle 3 

```python
# Problem 4.3
job3 = execute(deutsch_circuit3, backend, shots =1)
result3 = job3.result()
sv3 = result3.get_statevector()
np.around(sv3,2)
```

```python
plot_bloch_multivector(sv3)
```

The third oracle is also balanced and we get the following statevector:

$$
\frac{1}{\sqrt{2}}\begin{bmatrix}0&-1\\0&1\end{bmatrix}=\frac{1}{\sqrt{2}}\Big(-\big|01\big\rangle+\big|11\big\rangle\Big)=\frac{1}{\sqrt{2}}\Big(-\big|0\big\rangle+\big|1\big\rangle\Big)\big|1\big\rangle
$$

We can see in the bloch vector visualization that qubit 0's state has been flipped from the input state $\big|0\big\rangle$ to the state $\big|1\big\rangle$ as with the other balanced circuit.

### 6.2 - Building Constant and Balanced Deutsch-Jozsa Circuits

```python
# Problem 6.2a - Contstant Oracle Deutsch-Jozsa Circuit
djozsa_circuit1 = QuantumCircuit(3,2)
djozsa_circuit1.reset([0,1,2])
djozsa_circuit1.x([2])
djozsa_circuit1.barrier()
djozsa_circuit1.h([0,1,2])
djozsa_circuit1.append(circuit5, [0,1,2])
djozsa_circuit1.h([0,1])
djozsa_circuit1.barrier()
djozsa_circuit1.measure([0,1],[0,1])
djozsa_circuit1.draw('mpl')
```

---

```python
# Problem 6.2b - Balanced Oracle Deutsch-Jozsa Circuit
djozsa_circuit2 = QuantumCircuit(3,2)
djozsa_circuit2.reset([0,1,2])
djozsa_circuit2.x([2])
djozsa_circuit2.barrier()
djozsa_circuit2.h([0,1,2])
djozsa_circuit2.append(circuit6, [0,1,2])
djozsa_circuit2.h([0,1])
djozsa_circuit2.barrier()
djozsa_circuit2.measure([0,1],[0,1])
djozsa_circuit2.draw('mpl')
```

### Problem 7
---

### 7.1 - Deutsch-Jozsa Statevector Simulaton - Constant Oracle

```python
# Problem 7.1 Deutsch-Jozsa Statevector Simulaton - Constant Oracle 
job5 = execute(djozsa_circuit1, backend, shots =1)
result5 = job5.result()
sv5 = result5.get_statevector()
np.around(sv5,2)
```

```python
plot_bloch_multivector(sv5)
```

As we can see from the array and Bloch sphere statevector representations above, the first two qubits are in the $\big|0\big\rangle$ state and the third is once again in the $\big|0\big\rangle-\big|1\big\rangle$ state. The statevector array we get back now includes values for eight basis states, though only two are non-zero: the  000 and the 100 basis states. So for this balanced circuit our statevector looks like this:

$$
\frac{1}{\sqrt{2}}\Big(\big|000\big\rangle-\big|100\big\rangle\Big)=\frac{1}{\sqrt{2}}\Big(\big|0\big\rangle-\big|1\big\rangle\Big)\big|00\big\rangle
$$

### 7.2 - Deutsch-Jozsa Statevector Simulaton - Balanced Oracle

```python
# Problem 7.2 Deutsch-Jozsa Statevector Simulaton - Balanced Oracle 
job6 = execute(djozsa_circuit2, backend, shots =1)
result6 = job6.result()
sv6 = result6.get_statevector()
np.around(sv6,2)
```

```python
plot_bloch_multivector(sv6)
```

And here we see for the balance oracle Deutsch-Jozsa circuit we see that the top two qubits are in the down state while the third quibit is in it's expected $\big|0\big\rangle-\big|1\big\rangle$ state. The statevector for the system is confirmed balanced in the math as with the other examples: 

$$
\frac{1}{\sqrt{2}}\Big(\big|011\big\rangle-\big|111\big\rangle\Big)=\frac{1}{\sqrt{2}}\Big(\big|0\big\rangle-\big|1\big\rangle\Big)\big|11\big\rangle
$$

Again, we did not need to take measurements but only to probe the circuits in one shot in order to find whether the circuit was constant or balanced.
